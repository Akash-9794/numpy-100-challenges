```python
# ================================================
# NUMPY NOTES — dtype Deep Dive
# ================================================

import numpy as np

# ── ALL IMPORTANT DTYPES ───────────────────────

# INTEGER TYPES
i8  = np.array([1,2,3], dtype=np.int8)    # -128 to 127
i16 = np.array([1,2,3], dtype=np.int16)   # -32768 to 32767
i32 = np.array([1,2,3], dtype=np.int32)   # -2B to 2B
i64 = np.array([1,2,3], dtype=np.int64)   # very large!

# FLOAT TYPES
f16 = np.array([1.0], dtype=np.float16)   # low precision
f32 = np.array([1.0], dtype=np.float32)   # ML STANDARD ⭐
f64 = np.array([1.0], dtype=np.float64)   # default, heavy

# UNSIGNED (no negatives)
u8  = np.array([255], dtype=np.uint8)     # images! 0-255

# BOOLEAN
b   = np.array([1,0,1], dtype=bool)       # True/False

# ── MEMORY COMPARISON ─────────────────────────
print("Memory per element:")
print(f"int8   : {np.dtype(np.int8).itemsize} bytes")
print(f"int32  : {np.dtype(np.int32).itemsize} bytes")
print(f"int64  : {np.dtype(np.int64).itemsize} bytes")
print(f"float16: {np.dtype(np.float16).itemsize} bytes")
print(f"float32: {np.dtype(np.float32).itemsize} bytes")
print(f"float64: {np.dtype(np.float64).itemsize} bytes")

# ── CONVERSION ────────────────────────────────
arr = np.array([1.7, 2.9, 3.1])

# Method 1: astype()
int_arr = arr.astype(np.int32)    # [1 2 3] truncates!
f32_arr = arr.astype(np.float32)

print(f"\nOriginal : {arr}")
print(f"To int32 : {int_arr}")    # decimals LOST!
print(f"To float32: {f32_arr}")

# ── DANGER ZONE ───────────────────────────────
# Overflow! What happens when value exceeds range?
big = np.array([300], dtype=np.int32)
small = big.astype(np.int8)     # int8 max = 127!
print(f"\nOverflow: {small}")    # WRONG value! 😱

# ── ML STANDARD ───────────────────────────────
# Images pipeline:
img = np.random.randint(0, 255, (28,28), dtype=np.uint8)
print(f"\nImage dtype  : {img.dtype}")    # uint8

# Normalize for ML:
img_normalized = img.astype(np.float32) / 255.0
print(f"Normalized   : {img_normalized.dtype}")  # float32
print(f"Value range  : {img_normalized.min():.2f} to {img_normalized.max():.2f}")
```
