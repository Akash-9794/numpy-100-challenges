````python
import numpy as np

# Q1: Create 8 evenly spaced numbers
#     from 0 to 7 using arange

arr1 = np.arange(0,8)
# print(arr1)

# Q2: Create 8 evenly spaced numbers
#     from 0 to 7 using linspace
#     (notice the difference in output!)

arr2 = np.linspace(0,7,8)
# print(arr2)

# Q3: Create learning rate schedule
#     50 values from 0.01 to 0.0001

arr3 = np.linspace(0.01, 0.0001, 50)
# print(arr3)

# Q4: Create batch indices for:
#     dataset of 500 samples, batch size 50
dataset_size = 500
batch_size = 50
arr4 = np.arange(0,500,50)
# print(arr4)

# Q5: Tricky! 🔥
#     What is the difference in OUTPUT between:
#     np.arange(0, 1, 0.1)
#     np.linspace(0, 1, 10)
#     Write both AND explain difference!

arr5 = np.arange(0,1,0.1)
arr6 = np.linspace(0,1,10)
# print(arr5)
# print(arr6)

## Simple One Line to Remember:
# ```
# arange  → keeps adding step  → errors accumulate → unpredictable!
# linspace → calculates directly → no accumulation  → always safe! ✅
# ```
````
