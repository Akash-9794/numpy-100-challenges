#Topic: Data Types (dtype)
This is important because: → Wrong dtype = slow ML training → Wrong dtype = memory waste → Every ML engineer must know this!



```python
# Always use these in ML:
np.int32      # integers (saves memory vs int64)
np.float32    # floats (ML training standard!)
bool          # True/False (no np. prefix!)
np.uint8      # images (0-255 pixel values)

# Avoid in ML:
np.float64    # too heavy! (default but wasteful)
np.int64      # too heavy for most cases!
```
