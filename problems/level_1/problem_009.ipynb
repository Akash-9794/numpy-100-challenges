#Problem #009 Ready! 🔥
Topic:  Random Arrays!

```python
# ================================================
# NUMPY NOTES — Random Module
# ================================================

import numpy as np

# ── 5 MOST IMPORTANT RANDOM FUNCTIONS ─────────

# 1. rand() → uniform distribution 0 to 1
r1 = np.random.rand(5)          # 1D
r2 = np.random.rand(3,3)        # 2D matrix

# 2. randn() → normal distribution (mean=0, std=1)
n1 = np.random.randn(5)         # values around 0
n2 = np.random.randn(3,3)       # 2D normal

# 3. randint() → random integers
i1 = np.random.randint(0, 10, 5)        # 5 ints 0-9
i2 = np.random.randint(0, 255, (3,3))   # 2D matrix

# 4. seed() → reproducibility! ⭐ ML CRITICAL!
np.random.seed(42)              # fix randomness
a = np.random.rand(5)
np.random.seed(42)              # reset seed
b = np.random.rand(5)
print(np.array_equal(a, b))     # True! same values!

# 5. choice() → random selection
arr = np.array([10,20,30,40,50])
c1 = np.random.choice(arr, 3)           # pick 3
c2 = np.random.choice(arr, 3, replace=False)  # no repeat

# ── rand() vs randn() ─────────────────────────
print("rand()  → always 0 to 1:")
print(np.random.rand(5))

print("\nrandn() → can be negative!")
print(np.random.randn(5))
# [-1.2, 0.3, 2.1, -0.5, 0.8] → around 0!

# ── seed() WHY IMPORTANT IN ML ────────────────
# Without seed → different results every run!
# With seed    → same results every run!

# Example: train/test split must be reproducible!
np.random.seed(42)
data = np.random.rand(100, 10)
# Everyone who runs this gets SAME data! ✅

# ── ML Weight Initialization ──────────────────
np.random.seed(0)
# Xavier initialization (popular in deep learning!)
input_size  = 784
output_size = 128
W = np.random.randn(input_size, output_size) * np.sqrt(2/input_size)
print(f"\nWeights shape : {W.shape}")
print(f"Weights mean  : {W.mean():.4f}")   # close to 0!
print(f"Weights std   : {W.std():.4f}")    # close to 1!
```
