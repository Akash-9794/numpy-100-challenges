```python
# ================================================
# NUMPY NOTES — ML Indexing Patterns
# ================================================

import numpy as np

# ── PATTERN 1: TRAIN/TEST SPLIT ───────────────
np.random.seed(42)
dataset = np.random.rand(100, 5)
labels  = np.random.randint(0, 2, 100)

# 80/20 split
split = int(0.8 * len(dataset))   # 80
train_X = dataset[:split].copy()
test_X  = dataset[split:].copy()
train_y = labels[:split].copy()
test_y  = labels[split:].copy()

print(f"Train: {train_X.shape}")  # (80,5)
print(f"Test : {test_X.shape}")   # (20,5)

# ── PATTERN 2: MINI BATCH ─────────────────────
def get_batch(X, y, batch_size, batch_num):
    start = batch_num * batch_size
    end   = start + batch_size
    return X[start:end], y[start:end]

batch_X, batch_y = get_batch(train_X, train_y, 32, 0)
print(f"Batch shape: {batch_X.shape}")  # (32,5)

# ── PATTERN 3: FEATURE SELECTION ──────────────
# Select important features by index
important = [0, 2, 4]
X_selected = dataset[:, important]
print(f"Selected: {X_selected.shape}")  # (100,3)

# ── PATTERN 4: CLASS FILTERING ────────────────
# Get samples of specific class
class0 = dataset[labels == 0]
class1 = dataset[labels == 1]
print(f"Class 0: {class0.shape}")
print(f"Class 1: {class1.shape}")

# ── PATTERN 5: NORMALIZATION ──────────────────
# Normalize each feature (column) separately
X = np.random.rand(100, 4)
X_min = X.min(axis=0)    # min of each column
X_max = X.max(axis=0)    # max of each column
X_norm = (X - X_min) / (X_max - X_min)
print(f"Normalized range: {X_norm.min():.2f} to {X_norm.max():.2f}")

# ── PATTERN 6: SHUFFLE DATASET ────────────────
np.random.seed(42)
indices = np.random.permutation(len(dataset))
dataset_shuffled = dataset[indices]
labels_shuffled  = labels[indices]
print(f"Shuffled shape: {dataset_shuffled.shape}")

# ── PATTERN 7: TOP-K PREDICTIONS ──────────────
# Get indices of top 3 predictions
probs = np.array([0.1, 0.8, 0.3, 0.9, 0.5])
top_k = np.argsort(probs)[-3:][::-1]  # top 3 indices
print(f"Top 3 indices: {top_k}")       # [3,1,4]
print(f"Top 3 values : {probs[top_k]}")# [0.9,0.8,0.5]
```

#Solution


```python
import numpy as np
np.random.seed(42)

dataset = np.random.rand(200, 8)
labels  = np.random.randint(0, 3, 200)  # 3 classes!

# Q1: Split dataset 70/30 train/test
#     Use .copy() for safety!
#     Print both shapes
split = int(0.7* len(dataset))
train = dataset[:split].copy()
test = dataset[split:].copy()
print(train.shape)
print(test.shape)

# Q2: Get mini batch
#     batch_size = 16
#     Get 3rd batch (batch_num = 2)
#     Print shape

def get_batch(x, batch_size, batch_num):
  start = batch_num * batch_size
  end = start + batch_size
  return x[start:end]

batch_x = get_batch(train , 16, 2)
print(f"Batch shape: {batch_x.shape}")

# Q3: Feature selection
#     Select features at index 1,3,5,7
#     Print shape

imp = [1,3,5,7]
x_selected = dataset[:,imp]
print(f"Batch shape: {x_selected.shape}")

# Q4: Class filtering
#     Get all samples of class 2 only
#     Print shape

class2 = dataset[labels == 2]
print(f"class 2: {class2.shape}")

# Q5: ML Pipeline! 🔥🔥
#     a) Shuffle dataset with seed=99
#     b) Split 80/20
#     c) Select first 4 features only
#     d) Get all class 0 samples from training set
#     e) Print shape at each step!

# Q5 — Full ML Pipeline
# Step by step hints:

# a) Shuffle with seed=99
np.random.seed(99)
indices = np.random.permutation(len(dataset))  # length of dataset
dataset_s = dataset[indices]             # apply indices
labels_s  = labels[indices]              # shuffle labels too!
print(f"a) Shuffled: {dataset_s.shape}")

# b) Split 80/20
split = int(0.8 * len(dataset))
train_X = dataset_s[:split].copy()
train_y = labels_s[:split].copy()
print(f"b) Train: {train_X.shape}")

# c) First 4 features only
train_X4 = train_X[:, :4]
print(f"c) Features: {train_X4.shape}")

# d) Class 0 from training set
class0 = train_X4[train_y == 0]
print(f"d) Class 0: {class0.shape}")
```
