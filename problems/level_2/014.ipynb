#Problem #014 —  np.where() :

```python
import numpy as np

scores = np.array([0.9, 0.3, 0.7, 0.4, 0.8, 0.2, 0.6])

# Q1: Convert to binary
#     >= 0.5 → 1
#     <  0.5 → 0

# Q2: Replace values < 0.5 with 0.5
#     (minimum threshold)

# Q3: Find INDICES of high confidence
#     predictions (scores > 0.7)

# Q4: ML Task! 🔥
labels = np.array(['spam','ham','spam','ham','spam'])
#     Encode: spam=1, ham=0
#     Use np.where()
```

#Solution


```python
import numpy as np

scores = np.array([0.9, 0.3, 0.7, 0.4, 0.8, 0.2, 0.6])

# Q1: Convert to binary
#     >= 0.5 → 1
#     <  0.5 → 0

binary = np.where(scores >= 0.5 , 1,0)
print(binary)

# Q2: Replace values < 0.5 with 0.5
#     (minimum threshold)

min_threshold = np.where(scores < 0.5,0.5,scores)
print(min_threshold)
# Q3: Find INDICES of high confidence
#     predictions (scores > 0.7)

indices = np.where(scores > 0.7)
print(indices)

# Q4: ML Task! 🔥
labels = np.array(['spam','ham','spam','ham','spam'])
#     Encode: spam=1, ham=0
#     Use np.where()

encode = np.where(labels == "spam",1,0)
print(encode)
```
