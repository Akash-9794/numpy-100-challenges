#Problem #011 — 1D Indexing:

```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

# Q1: Get element at index 3

# Q2: Get last element without hardcoding index

# Q3: Get elements from index 2 to 7

# Q4: Get every 3rd element

# Q5: Reverse the array

# Q6: Get last 4 elements

# Q7: ML Task!
#     Split into 2 equal batches
#     batch1 = first 5 elements
#     batch2 = last 5 elements
```

#Solution


```python
import numpy as np

arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

# Q1: Get element at index 3
print(arr[3])
# Q2: Get last element without hardcoding index
print(arr[-1])
# Q3: Get elements from index 2 to 7
print(arr[2:8])
# Q4: Get every 3rd element
print(arr[::3])
# Q5: Reverse the array
print(arr[::-1])
# Q6: Get last 4 elements
print(arr[-4:])
# Q7: ML Task!
#     Split into 2 equal batches
#     batch1 = first 5 elements
#     batch2 = last 5 elements

batch1 = arr[:5]
batch2 = arr[5:]

print(batch1)
print(batch2)
```
