#Problem #013 — Boolean Masking :

```python
import numpy as np

data = np.array([0.1, 0.8, 0.3, 1.5, 0.6, -0.2, 0.9, 2.1])

# Q1: Get all values greater than 0.5

# Q2: Get all values between 0 and 1

# Q3: Replace all values outside 0-1 with -1

# Q4: Count how many values are greater than 0.5
#     Hint: True=1, False=0 → use sum()!

# Q5: ML Task! 🔥
#     These are model predictions (probabilities)
#     Values should be between 0 and 1
#     Find and print INVALID predictions
#     (less than 0 OR greater than 1)
```

#Solution


```python
import numpy as np

data = np.array([0.1, 0.8, 0.3, 1.5, 0.6, -0.2, 0.9, 2.1])

# Q1: Get all values greater than 0.5

print(data[data > 0.5])

# Q2: Get all values between 0 and 1

print(data[(data > 0) & (data < 1)])

# Q3: Replace all values outside 0-1 with -1
datac = np.copy(data)
datac[(datac < 0) | (datac > 1)] = -1
print(datac)

# Q4: Count how many values are greater than 0.5
#     Hint: True=1, False=0 → use sum()!

arr1 = (data > 0.5).sum()
print(arr1)

# Q5: ML Task! 🔥
#     These are model predictions (probabilities)
#     Values should be between 0 and 1
#     Find and print INVALID predictions
#     (less than 0 OR greater than 1)

invalid = data[((data < 0) | (data > 1))]
print(invalid)
```
