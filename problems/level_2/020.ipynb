#Solution


```python
import numpy as np
np.random.seed(42)

# Dataset: 500 samples, 10 features, 4 classes
X = np.random.rand(500, 10)
y = np.random.randint(0, 4, 500)

# ─────────────────────────────────────────────
# CHALLENGE 1: Complete ML Pipeline
# ─────────────────────────────────────────────
# a) Shuffle X and y together (seed=42)
np.random.seed(42)
indices = np.random.permutation(len(X))
X_s = X[indices]
y_s = y[indices]

# b) 70/15/15 train/val/test split

train_end = int(0.7 * len(X_s))
val_end = int(0.85 * len(X_s))

X_train = X_s[:train_end].copy()
X_val = X_s[train_end : val_end].copy()
X_test = X_s[val_end:].copy()

# c) Print all 3 shapes

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)


# ─────────────────────────────────────────────
# CHALLENGE 2: Feature Engineering
# ─────────────────────────────────────────────
# a) From train set select odd-indexed features
#    (features 1,3,5,7,9)

odd_features = X_train[:,1:10:2]

# b) From train set select even-indexed features
#    (features 0,2,4,6,8)

even_features = X_train[:,0:10:2]

# c) Print both shapes

print(odd_features.shape)
print(even_features.shape)

# ─────────────────────────────────────────────
# CHALLENGE 3: Class Analysis
# ─────────────────────────────────────────────
# a) Count samples per class in training set
#    Hint: use loop + boolean mask + sum()
# b) Find which class has most samples
#    Hint: np.argmax()

y_train = y_s[:train_end].copy()

for class_num in range(4):
  count = (y_train == class_num).sum()
  print(f"Class {class_num}: {count} samples")

counts = np.bincount(y_train)          # pass y_train
most_common = np.argmax(counts)       # pass counts
print(f"Most common class: {most_common}")

# ─────────────────────────────────────────────
# CHALLENGE 4: Batch Generator
# ─────────────────────────────────────────────
# Create function get_batch(X, y, batch_size, batch_num)
# Get batches 0,1,2 from training set
# batch_size = 32
# Print shape of each batch

def get_batch(X, y, batch_size, batch_num):
  start = batch_num * batch_size
  end = start + batch_size
  return X[start:end],y[start:end]

for i in range(3):
  batch_X , batch_y = get_batch(X_train, y_train, 32,i)
  print(f"Batch {i}: {batch_X.shape}")

# ─────────────────────────────────────────────
# CHALLENGE 5: Advanced Filtering 🔥🔥
# ─────────────────────────────────────────────
# From test set:
# a) Get samples where ALL features > 0.5
#    Hint: use axis parameter in np.all()

mask_all = np.all(X_test > 0.5, axis=1)
result_all = X_test[mask_all]
print(f"All > 0.5: {result_all.shape}")

# b) Get samples where ANY feature > 0.9
#    Hint: use axis parameter in np.any()
# c) Print both shapes

mask_any = np.any(X_test > 0.9 , axis=1)
result_any = X_test[mask_any]
print(f"any > 0.9 {result_any.shape}")
```
