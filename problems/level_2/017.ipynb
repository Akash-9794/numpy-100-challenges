```python
# ================================================
# NUMPY NOTES — Views vs Copies
# ================================================

import numpy as np

# ── THE PROBLEM ───────────────────────────────
# NumPy slicing returns VIEW not copy!
# Modifying view → modifies original! 😱

original = np.array([1, 2, 3, 4, 5])

# SLICE → returns VIEW
view = original[1:4]
print(view)          # [2 3 4]

view[0] = 99         # modify view
print(original)      # [1 99 3 4 5] ← CHANGED! 😱

# ── HOW TO MAKE REAL COPY ─────────────────────
original2 = np.array([1, 2, 3, 4, 5])

# Method 1: .copy()
copy = original2[1:4].copy()
copy[0] = 99
print(original2)     # [1 2 3 4 5] ← SAFE! ✅

# Method 2: np.copy()
copy2 = np.copy(original2[1:4])

# ── CHECK IF VIEW OR COPY ─────────────────────
arr = np.array([1, 2, 3, 4, 5])
view = arr[1:4]
copy = arr[1:4].copy()

print(view.base is arr)   # True  → it's a VIEW!
print(copy.base is arr)   # False → it's a COPY!

# ── FANCY INDEXING → ALWAYS COPY! ─────────────
arr2 = np.array([10, 20, 30, 40, 50])

fancy = arr2[[0, 2, 4]]   # fancy indexing
fancy[0] = 999
print(arr2)    # [10 20 30 40 50] ← unchanged! ✅
# Fancy indexing always returns COPY!

# ── BOOLEAN MASK → ALWAYS COPY! ───────────────
mask_result = arr2[arr2 > 15]
mask_result[0] = 999
print(arr2)    # unchanged ✅

# ── SUMMARY ───────────────────────────────────
# Slicing    → VIEW   (modifies original!) ⚠️
# .copy()    → COPY   (safe!) ✅
# Fancy idx  → COPY   (safe!) ✅
# Bool mask  → COPY   (safe!) ✅

# ── ML CONTEXT ────────────────────────────────
# DANGER in ML preprocessing!
dataset = np.random.rand(100, 5)

# Wrong way:
train = dataset[:80]      # VIEW! ⚠️
train[:] = 0              # accidentally zeros dataset!

# Right way:
train = dataset[:80].copy()   # COPY! ✅
train[:] = 0              # dataset safe!
print(dataset[0])         # original unchanged ✅
```

#Solution


```python
# ================================================
# Problem #017 — Views vs Copies
# ================================================
# KEY CONCEPT:
# NumPy slicing returns a VIEW (not a copy!)
# VIEW = shares same memory as original array
# COPY = independent memory, original stays safe
#
# Rule:
# Slicing       → VIEW  (dangerous! modifies original)
# Fancy indexing → COPY (safe! original unchanged)
# Boolean mask  → COPY (safe! original unchanged)
# .copy()       → COPY (always safe!)
# ================================================

import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])

# ── Q1: VIEW ──────────────────────────────────
# Slicing returns VIEW → shares memory with original
# Modifying view WILL modify original array!

view = arr[2:7]
view[0] = 999          # modifying view
print(f"View    : {view}")
print(f"Original: {arr}")
# Original CHANGES because view shares same memory!
# arr[2] becomes 999 automatically

# ── Q2: COPY ──────────────────────────────────
# .copy() creates independent memory
# Modifying copy will NOT affect original array!

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])  # reset
copy_ele = arr[2:7].copy()
copy_ele[0] = 999      # modifying copy
print(f"\nCopy    : {copy_ele}")
print(f"Original: {arr}")
# Original UNCHANGED because copy has its own memory!

# ── Q3: CHECK VIEW OR COPY ────────────────────
# .base tells us if array shares memory with another
# .base is arr → True  = VIEW (shares memory)
# .base is arr → False = COPY (independent memory)

a = arr[1:5]           # slicing → VIEW
b = arr[[1,2,3,4]]     # fancy indexing → COPY

print(f"\na.base is arr : {a.base is arr}")
# True → slicing always returns VIEW!

print(f"b.base is arr : {b.base is arr}")
# False → fancy indexing always returns COPY!

# ── Q4: ML TASK ───────────────────────────────
# This is a REAL ML bug that beginners make!
# When splitting dataset into train/test
# always use .copy() to avoid accidental changes!

np.random.seed(42)
dataset = np.random.rand(10, 4)

# ❌ WRONG way — train_wrong is a VIEW!
# Modifying train_wrong will change dataset!
train_wrong = dataset[:7]       # VIEW!
train_wrong[0,0] = 9999
print(f"\nWRONG — Dataset changed : {dataset[0,0]}")
# Prints 9999 — original dataset corrupted! 😱

# ✅ RIGHT way — train_right is a COPY!
# Modifying train_right will NOT change dataset2!
np.random.seed(42)
dataset2 = np.random.rand(10, 4)
train_right = dataset2[:7].copy()   # COPY!
train_right[0,0] = 9999
print(f"RIGHT  — Dataset safe   : {dataset2[0,0]:.4f}")
# Prints original value — dataset2 safe! ✅

# ── GOLDEN RULE ───────────────────────────────
# In ML ALWAYS use .copy() when splitting data:
# train = dataset[:80].copy()  ✅ safe!
# test  = dataset[80:].copy()  ✅ safe!
```
