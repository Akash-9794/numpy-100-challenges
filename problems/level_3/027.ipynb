```python
# ================================================
# NOTES — Universal Functions (ufuncs)
# ================================================

import numpy as np

# ── WHAT ARE UFUNCS? ──────────────────────────
# Built-in NumPy functions that:
# → operate element-wise on arrays
# → much faster than Python loops
# → work on any shape array!

# ── MATH UFUNCS ───────────────────────────────
arr = np.array([1.0, 4.0, 9.0, 16.0, 25.0])

print(np.sqrt(arr))      # [1. 2. 3. 4. 5.]
print(np.square(arr))    # [1. 16. 81. 256. 625.]
print(np.abs(arr))       # absolute values
print(np.exp(arr))       # e^x
print(np.log(arr))       # natural log ln(x)
print(np.log2(arr))      # log base 2
print(np.log10(arr))     # log base 10
print(np.power(arr, 3))  # arr^3

# ── ROUNDING ──────────────────────────────────
arr = np.array([1.2, 2.7, 3.5, 4.9, -1.3])

print(np.floor(arr))   # [1.  2.  3.  4. -2.] round DOWN
print(np.ceil(arr))    # [2.  3.  4.  5. -1.] round UP
print(np.round(arr))   # [1.  3.  4.  5. -1.] nearest
print(np.trunc(arr))   # [1.  2.  3.  4. -1.] remove decimal

# ── COMPARISON ────────────────────────────────
a = np.array([1, 5, 3, 8, 2])
b = np.array([4, 2, 3, 1, 6])

print(np.maximum(a, b))   # [4,5,3,8,6] element-wise max
print(np.minimum(a, b))   # [1,2,3,1,2] element-wise min
print(np.greater(a, b))   # [F,T,F,T,F] a>b

# ── ML ACTIVATION FUNCTIONS ───────────────────
# All built using ufuncs!

# Sigmoid: 1/(1+e^-x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU: max(0,x)
def relu(x):
    return np.maximum(0, x)

# Tanh: built-in ufunc!
def tanh(x):
    return np.tanh(x)

# Softmax: e^x / sum(e^x)
def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability trick!
    return e_x / e_x.sum()

# Leaky ReLU: max(0.01x, x)
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# ── CROSS ENTROPY LOSS ────────────────────────
# Uses np.log() ufunc!
def binary_cross_entropy(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1-1e-7)  # avoid log(0)!
    return -np.mean(y_true*np.log(y_pred) +
                   (1-y_true)*np.log(1-y_pred))

y_true = np.array([1, 0, 1, 0, 1])
y_pred = np.array([0.9, 0.1, 0.8, 0.2, 0.7])
loss = binary_cross_entropy(y_true, y_pred)
print(f"BCE Loss: {loss:.4f}")

# ── cumsum & cumprod ──────────────────────────
arr = np.array([1, 2, 3, 4, 5])

print(np.cumsum(arr))    # [1,3,6,10,15]
print(np.cumprod(arr))   # [1,2,6,24,120]
```

#Solution


```python
import numpy as np

arr = np.array([-4.0, -2.5, 0.0, 1.5, 3.7, 5.2])

# Q1: Apply sqrt to positive values only
#     Hint: use np.where + np.sqrt
positive = np.where(arr > 0)
print(np.sqrt(arr[positive]))
# Q2: Round each value using:
#     a) floor  b) ceil  c) round

print(np.floor(arr))
print(np.ceil(arr))
print(np.round(arr))

# Q3: Implement these activation functions:
z = np.array([-3.0, -1.0, 0.0, 1.0, 3.0])
#     a) sigmoid
#     b) relu
#     c) leaky relu (alpha=0.01)
#     d) tanh
# Sigmoid: 1/(1+e^-x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU: max(0,x)
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU: max(0.01x, x)
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Tanh: built-in ufunc!
def tanh(x):
    return np.tanh(x)

print(f"Sigmoid    : {sigmoid(z).round(3)}")
print(f"ReLU       : {relu(z)}")
print(f"Leaky ReLU : {leaky_relu(z).round(3)}")
print(f"Tanh       : {tanh(z).round(3)}")

# Q4: Cross Entropy Loss! 🔥

y_true = np.array([1, 0, 1, 1, 0])
y_pred = np.array([0.9, 0.2, 0.7, 0.8, 0.1])
# BCE = -mean(y*log(p) + (1-y)*log(1-p))
# Hint: use np.clip first to avoid log(0)!
# Uses np.log() ufunc!
def binary_cross_entropy(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1-1e-7)  # avoid log(0)!
    return -np.mean(y_true*np.log(y_pred) +
                   (1-y_true)*np.log(1-y_pred))

loss = binary_cross_entropy(y_true, y_pred)
print(f"BCE Loss: {loss:.4f}")

# Q5: ML Task! 🔥🔥
# Implement softmax for these logits:
logits = np.array([2.0, 1.0, 0.5, 3.0, 1.5])
# Hint: stability trick → subtract max first!
# Then verify sum = 1.0

# Softmax: e^x / sum(e^x)
def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability trick!
    return e_x / e_x.sum()

result = softmax(logits)
print(result)
print(f"Sum = {result.sum().round()}")
```
