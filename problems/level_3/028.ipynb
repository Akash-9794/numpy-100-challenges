```python
# ================================================
# NOTES — ML Math Operations
# ================================================

import numpy as np

# ── 1. NORMALIZATION TECHNIQUES ───────────────

# Min-Max Normalization (0 to 1)
def minmax_normalize(X):
    return (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

# Z-Score Standardization (mean=0, std=1)
def standardize(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)

# ── 2. DISTANCE METRICS ───────────────────────
# Used in KNN, clustering!

# Euclidean Distance
def euclidean(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Manhattan Distance
def manhattan(a, b):
    return np.sum(np.abs(a - b))

# Cosine Similarity
def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# ── 3. LOSS FUNCTIONS ─────────────────────────

# MSE Loss
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# MAE Loss
def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

# Binary Cross Entropy
def bce(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1-1e-7)
    return -np.mean(y_true*np.log(y_pred) +
                   (1-y_true)*np.log(1-y_pred))

# ── 4. ACCURACY METRICS ───────────────────────

# Classification Accuracy
def accuracy(y_true, y_pred_probs):
    y_pred = (y_pred_probs >= 0.5).astype(int)
    return np.mean(y_true == y_pred)

# ── 5. GRADIENT DESCENT STEP ──────────────────
# Heart of ML training!

# dL/dW = (1/n) * X.T @ (predictions - y)
def compute_gradient(X, y, predictions):
    n = len(y)
    dW = (1/n) * X.T @ (predictions - y)
    db = np.mean(predictions - y)
    return dW, db

# Update weights:
# W = W - learning_rate * dW
def update_weights(W, b, dW, db, lr=0.01):
    W = W - lr * dW
    b = b - lr * db
    return W, b

# ── 6. LINEAR REGRESSION FROM SCRATCH ─────────
np.random.seed(42)
X = np.random.rand(100, 3)
y = X @ np.array([2.0, 3.0, 1.0]) + 0.5

# Initialize
W = np.zeros(3)
b = 0.0
lr = 0.01

# Training loop
for epoch in range(100):
    predictions = X @ W + b
    loss = mse(y, predictions)
    dW, db = compute_gradient(X, y, predictions)
    W, b = update_weights(W, b, dW, db, lr)
    if epoch % 20 == 0:
        print(f"Epoch {epoch}: Loss={loss:.4f}")
```
