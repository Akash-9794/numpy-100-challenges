```python
# ================================================
# NOTES — argmin, argmax, argsort
# ================================================

import numpy as np

# ── DIFFERENCE: min vs argmin ─────────────────
arr = np.array([30, 10, 50, 20, 40])

print(np.min(arr))     # 10  → actual VALUE
print(np.argmin(arr))  # 1   → INDEX of min value!

print(np.max(arr))     # 50  → actual VALUE
print(np.argmax(arr))  # 2   → INDEX of max value!

# ── 2D argmin/argmax ──────────────────────────
matrix = np.array([[3, 1, 4],
                   [1, 5, 9],
                   [2, 6, 5]])

# No axis → index in FLATTENED array
print(np.argmax(matrix))        # 5 → index in flat!

# axis=0 → index of max per COLUMN
print(np.argmax(matrix, axis=0))  # [0,2,1]
# col0: max=3 at row 0
# col1: max=6 at row 2
# col2: max=9 at row 1

# axis=1 → index of max per ROW
print(np.argmax(matrix, axis=1))  # [2,2,1]
# row0: max=4 at col 2
# row1: max=9 at col 2
# row2: max=6 at col 1

# ── argsort ───────────────────────────────────
arr = np.array([30, 10, 50, 20, 40])

# Returns INDICES that would sort array!
idx = np.argsort(arr)
print(idx)           # [1,3,0,4,2]
print(arr[idx])      # [10,20,30,40,50] → sorted! ✅

# Descending sort:
idx_desc = np.argsort(arr)[::-1]
print(arr[idx_desc]) # [50,40,30,20,10]

# ── TOP-K PATTERN ─────────────────────────────
# Most important ML pattern!
probs = np.array([0.1, 0.8, 0.3, 0.9, 0.5])

# Top 3 predictions:
top3_idx = np.argsort(probs)[-3:][::-1]
print(f"Top 3 indices: {top3_idx}")       # [3,1,4]
print(f"Top 3 values : {probs[top3_idx]}")# [0.9,0.8,0.5]

# ── ML CONTEXT ────────────────────────────────
# 1. Find best model from results
val_accuracies = np.array([0.82, 0.91, 0.87, 0.95, 0.88])
best_model = np.argmax(val_accuracies)
print(f"Best model index: {best_model}")    # 3
print(f"Best accuracy   : {val_accuracies[best_model]}") # 0.95

# 2. Rank features by importance
importance = np.array([0.3, 0.1, 0.5, 0.2, 0.4])
ranked_idx = np.argsort(importance)[::-1]
print(f"Feature ranking: {ranked_idx}")  # [2,4,0,3,1]

# 3. Find worst predictions
errors = np.abs(predictions - actual)
worst_idx = np.argsort(errors)[-3:][::-1]
print(f"Worst predictions at: {worst_idx}")
```

```python
import numpy as np
np.random.seed(42)

scores = np.array([0.3, 0.9, 0.1, 0.8, 0.5, 0.7, 0.2, 0.6])

# Q1: Find index of highest score
print(np.argmax(scores))
# Q2: Find index of lowest score
print(np.argmin(scores))
# Q3: Sort scores ascending → print sorted values
assending = np.argsort(scores)
print(scores[assending])
# Q4: Sort scores descending → print sorted values
print(scores[np.argsort(scores)[::-1]])
# Q5: Get top 3 scores and their indices
top_3scorIdx = assending[-3:][::-1]
print(top_3scorIdx)
print(scores[top_3scorIdx])
# Q6: 2D Challenge!
matrix = np.array([[0.8, 0.1, 0.3],
                   [0.2, 0.9, 0.4],
                   [0.5, 0.3, 0.7]])
# Find index of max per row (predicted class!)
# This is exactly what a classifier does!

idx_maxRow = np.argmax(matrix,axis=1)
print(idx_maxRow)

# Q7: ML Task! 🔥🔥
val_acc = np.array([0.75, 0.82, 0.79, 0.91, 0.88, 0.85])
# a) Which epoch had best accuracy?
# b) Which epoch had worst accuracy?
# c) Rank all epochs best to worst
# d) Print accuracies in ranked order

print(np.argmax(val_acc))
print(np.argmin(val_acc))
print(np.argsort(val_acc))
print(val_acc[np.argsort(val_acc)])
```
