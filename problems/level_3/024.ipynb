#Solution


```python
import numpy as np

np.random.seed(42)
dataset = np.random.rand(6, 4)
# 6 samples, 4 features

# Q1: Find sum, mean, std, min, max
#     of ENTIRE dataset

print(np.sum(dataset).round(5))
print(np.mean(dataset).round(5))
print(np.std(dataset).round(5))
print(np.min(dataset))
print(np.max(dataset))

# Q2: Find mean of each FEATURE (column)
#     → which feature has highest mean?

feature = np.mean(dataset,axis=0)
print(np.argmax(feature))

# Q3: Find mean of each SAMPLE (row)
#     → which sample has highest mean?
sample = np.mean(dataset,axis=1)
print(np.argmax(dataset))
# Q4: Find min and max of each column

min = np.min(dataset,axis = 0)
max = np.max(dataset,axis = 0)
print(min)
print(max)


# Q5: ML Task! 🔥
#     Find range (max-min) of each feature
#     This tells you how spread out each feature is!

range_per_feature = np.max(dataset, axis=0) - np.min(dataset,axis=0)
print(f"Range per feature: {range_per_feature.round(4)}")

# Q6: ML Task! 🔥🔥
labels = np.array([0, 1, 0, 1, 1, 0])
#     Count samples per class
#     Hint: use sum() with boolean mask!
#     Print class 0 count and class 1 count

class0_count = (labels == 0).sum()
class1_count = (labels == 1).sum()
print(f"Class 0:  {class0_count}")
print(f"Class 1:  {class1_count}")

```
