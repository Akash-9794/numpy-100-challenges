#Solution


```python
import numpy as np

# Q1: Dot product of two 1D arrays
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
# Calculate using np.dot() AND @ operator
# Are results same? -- yes

print(np.dot(a,b))
print(a @ b)

# Q2: Matrix multiplication
A = np.array([[1, 2, 3],
              [4, 5, 6]])    # shape (2,3)

B = np.array([[1, 2],
              [3, 4],
              [5, 6]])       # shape (3,2)
# Calculate A @ B → what shape?

print((A @ B).shape)

# Q3: Will this work?
#     A @ A  → why or why not?
# error ((A @ A).shape):  matmul: Input operand 1 has a mismatch in its core dimension 0, with gufunc signature (n?,k),(k,m?)->(n?,m?) (size 2 is different from 3)

print((A @ A).shape)

# Q4: Transpose trick! 🔥
#     Calculate A @ A.T → what shape?
#     Calculate A.T @ A → what shape?

print((A @ A.T).shape)
print((A.T @ A).shape)

# Q5: ML Task! 🔥🔥
#     Neural network forward pass!
np.random.seed(42)
X = np.random.rand(32, 784)    # 32 images, 784 pixels
W = np.random.rand(784, 128)   # weights
b = np.zeros(128)              # bias
# Calculate output = X @ W + b
# Print output shape
output = X @ W + b
print(output.shape)

# Q6: Linear Regression! 🔥🔥
np.random.seed(42)
X = np.random.rand(100, 3)           # 100 samples, 3 features
weights = np.array([0.5, 0.3, 0.2])  # 3 weights
bias = 0.1
# Calculate predictions = X @ weights + bias
# Print predictions shape

predictions = X @ weights + bias
print(predictions.shape)
```
