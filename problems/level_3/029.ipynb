```python
# ================================================
# NOTES — Full ML Pipeline Math
# ================================================

import numpy as np

# ── COMPLETE ML PIPELINE ──────────────────────
# Step 1: Load & prepare data
# Step 2: Normalize features
# Step 3: Initialize weights
# Step 4: Forward pass
# Step 5: Calculate loss
# Step 6: Backpropagation (gradients)
# Step 7: Update weights
# Step 8: Repeat steps 4-7

# ── TRAIN/VAL SPLIT ───────────────────────────
def train_val_split(X, y, val_size=0.2, seed=42):
    np.random.seed(seed)
    idx = np.random.permutation(len(X))
    split = int(len(X) * (1-val_size))
    return (X[idx[:split]], y[idx[:split]],
            X[idx[split:]], y[idx[split:]])

# ── CONFUSION MATRIX ──────────────────────────
def confusion_matrix(y_true, y_pred_probs, threshold=0.5):
    y_pred = (y_pred_probs >= threshold).astype(int)
    TP = np.sum((y_true==1) & (y_pred==1))
    TN = np.sum((y_true==0) & (y_pred==0))
    FP = np.sum((y_true==0) & (y_pred==1))
    FN = np.sum((y_true==1) & (y_pred==0))
    return TP, TN, FP, FN

# ── PRECISION, RECALL, F1 ─────────────────────
def precision(TP, FP):
    return TP / (TP + FP + 1e-7)

def recall(TP, FN):
    return TP / (TP + FN + 1e-7)

def f1_score(prec, rec):
    return 2 * (prec * rec) / (prec + rec + 1e-7)
```

```python
import numpy as np
np.random.seed(42)

# Dataset
X = np.random.rand(100, 4)
y = (X[:, 0] + X[:, 1] > 1.0).astype(float)

# ── STEP 1: Normalize ─────────────────────────
# Apply min-max normalization on X
# YOUR CODE:
X_norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))


# ── STEP 2: Train/Val Split ───────────────────
# 80/20 split manually
# YOUR CODE:
split = int(0.8 * len(X_norm))
X_train, X_val = X_norm[:split], X_norm[split:]
y_train, y_val = y[:split], y[split:]
print(f"Train: {X_train.shape}, Val: {X_val.shape}")

print(f"Train: {X_train.shape}")
print(f"Val  : {X_val.shape}")

# ── STEP 3: Initialize ────────────────────────
# weights = zeros shape (4,)
# bias = 0.0
# YOUR CODE:
W = np.zeros(4)
b = 0.0

# ── STEP 4: Training Loop ─────────────────────
# Train 100 epochs, lr=0.01
# Each epoch:
# → predictions = sigmoid(X_train @ W + b)
# → loss = BCE
# → gradients = compute_gradient
# → update weights
# Print loss every 20 epochs
# YOUR CODE:
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def bce(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1-1e-7)
    return -np.mean(y_true*np.log(y_pred) +
                   (1-y_true)*np.log(1-y_pred))

def compute_gradient(X, y, predictions):
    n = len(y)
    dW = (1/n) * X.T @ (predictions - y)
    db = np.mean(predictions - y)
    return dW, db

for epoch in range(100):
    # forward pass
    predictions = sigmoid(X_train @ W + b)
    # loss
    loss = bce(y_train, predictions)
    # gradients
    dW, db = compute_gradient(X_train, y_train, predictions)
    # update
    W = W - 0.01 * dW
    b = b - 0.01 * db
    if epoch % 20 == 0:
        print(f"Epoch {epoch}: Loss={loss:.4f}")

# ── STEP 5: Evaluate on Validation ────────────
# predictions on X_val
# calculate accuracy
# YOUR CODE:
val_preds = sigmoid(X_val @ W + b)
val_acc = np.mean((val_preds >= 0.5) == y_val)
print(f"Val Accuracy: {val_acc:.4f}")
```
